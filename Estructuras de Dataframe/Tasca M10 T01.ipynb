```python
# Inicializamos el navegador
URL = "http://quotes.toscrape.com"
driver.get(URL )
```

```python

frase = []

for web in driver.find_elements(By.CLASS_NAME,'quote'):
        #print(web.text)
        #data = [item.text for item in web.find_elements(By.CLASS_NAME, value= "text")]
        datafrase = [item.text for item in web.find_elements(By.CSS_SELECTOR,'span.text')]
        dataauthor = [item.text for item in web.find_elements(By.CSS_SELECTOR,'small.author')]
        links = [elem.get_attribute('href') for elem in web.find_elements(By.CSS_SELECTOR,"a")]
            
        frase.append([(''.join(datafrase)).replace('“','').replace('”',''),''.join(dataauthor),links[0]])
       
```

```python
df_Quotes_Sl=pd.DataFrame(frase,columns=['Frases','Author','Web'])
HTML(df_Quotes_Sl.to_html(render_links=True, escape=False))
```

### <font color='#d50283'>- Exercici 3.</font>
Tria una pàgina web que tu vulguis i realitza web scraping mitjançant la llibreria Selenium primer i Scrapy després. 

```python
scrapydo.setup()
class QuotesSpider(scrapy.Spider):     #creo una classe spider
    name = "quotes"
    start_urls = ['http://quotes.toscrape.com/tag/humor/']
    
    def parse(self, response):         #funció que "parseja" el contingut de la pàgina web
        lnk ='http://quotes.toscrape.com'
        for quote in response.css('div.quote'):
            print('Frase: ',quote.css('span.text::text').extract_first())
            print('Author:',quote.css('small.author::text').extract_first())#.extract_second())
            print("link: ",lnk+quote.css('a::attr(href)').extract_first())
            
            
            
            
scrapydo.run_spider(QuotesSpider)
```
