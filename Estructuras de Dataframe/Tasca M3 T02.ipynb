****Descripción de los campos del archivo:****

El archivo usado contiene la información sobre vuelos en el 2008. Descripción de los campos :

- ***Year:*** Año.
- ***Month:*** Mes.
- ***DayofMonth:*** Día del mes.
- ***DayOfWeek:*** 1 (Monday) - 7 (Sunday).
- ***DepTime:*** Hora de salida (local, hhmm).
- ***CRSDepTime:*** hora de salida programada (local, hhmm).
- ***AirTime:*** hora real de llegada (local, hhmm).
- ***CRSArrTime:*** hora de llegada prevista (local, hhmm).
- ***UniqueCarrier:*** código de operador único.
- ***FlightNum:*** Número de vuelo.

- ***ArrDelay:*** Diferencia en minutos entre la hora de llegada prevista y la real. Las llegadas anticipadas muestran números negativos, (en minutos).
- ***DepDelay:*** Diferencia en minutos entre la hora de salida prevista y la real. Las salidas anticipadas muestran números negativos., (en minutos).

- ***Dest:*** Destino del vuelo, código IATA de Aeropuertos.
- ***Distance:*** distancia en millas entre Aeropuertos.
- ***TaxiIn:*** Descenso de ruedas y llegada a la puerta del aeropuerto de destino, (en minutos).
- ***TaxiOut:*** El tiempo transcurrido entre la salida de la puerta del aeropuerto de origen y ruedas fuera (en minutos).
- ***Cancelled:*** se cancelo el vuelo?
- ***CancellationCode:*** Razones de cancelación (A = carrier, B = weather, C = NAS, D = security).
- ***Diverted:*** El avión ha sido desviado, 1 = yes, 0 = no.
- ***CarrierDelay:*** (en minutos).
- ***WeatherDelay:*** (en minutos).
- ***NASDelay:*** (en minutos).
- ***SecurityDelay:*** (en minutos).
- ***LateAircraftDelay:*** (en minutos).

****1.2.1 Eliminamos Registros del DataSet:****

Eliminamos los regitros dependiendo de la casuística descrita en el código
antes de reducir el DataSet.

```python
# Número de registros del dataSet antes de eliminar.
treg=len(data)
print('Resultado:\n ')
# eliminamos Filas con datos nulos
data.dropna(how='all')
if (treg==len(data)): 
    print('No tiene Filas Nulas.')
else:
    print(f'Total Filas nulas es: {treg-len(data)}.')
    
treg=len(data)

# eliminamos Filas duplicadas
data.drop_duplicates()

if treg==len(data): 
    print('No tiene Duplicados.')
else:
    print(f'Total de duplicados eliminados es: {treg-len(data)}.')

treg=len(data)

# En el pto del ejercicio  2.1 resumen estadístico para el campo AirTime el mínimo 
# es igual a 0. Volvemos a este pto. para eliminar estos registros

data.drop(data[data['AirTime']==0].index, inplace= True)

if treg==len(data): 
    print('No tiene vuelos realizados con AirTime=0')
else:
    print(f'Número Filas eliminadas con AirTime=0 son: {treg-len(data)}.')

```

****1.2.2 Reducir el DataSet:****
    
En este estudio nos piden reducir a 200000 registros, haremos este paso antes del estudio de null.

```python
# Reducimos el Dataset a 200000 registros y reset al index
airlines=data.sample(n=200000, frac=None, replace=False, weights=None).reset_index(drop = True)

#Aunque ya conocemos el número de registros, ampliamos la información
#con las columnas que tiene nuestro DataSet
print('Resultado:\n ')
print(f'Total de Registros: {airlines.shape[0]}')
```

****1.2.3 Creación y Eliminación de Variables:****

Se crea una columna nueva con la fecha y posteriormente eliminamos las tres columnas de Year, Month, Dayofmonth, el DayofWeek lo modificamos, pero lo dejamos para observar en las muestra que ¿día de la semana puede tener más retrasos? 



```python
DiaSemana = {1:'Lunes',2:'Martes',3:'Miércoles',4:'Jueves',5:'Viernes',6:'Sábado',7:'Domingo'}
airlines['DayOfWeek'] = airlines['DayOfWeek'].map(DiaSemana)
#Una forma de crear la variable, pero se va al final del dataframe
#airlines['Date']=airlines['DayofMonth'].astype('str')+'/'+airlines['Month'].astype('str')+'/'+airlines['Year'].astype('str')
airlines.insert(0,'Date',airlines['DayofMonth'].astype('str')+'/'+airlines['Month'].astype('str')+'/'+airlines['Year'].astype('str'))
airlines['Ruta']=airlines['Origin']+'-'+airlines['Origin']

#Observar el contenido del campo ()
#airlines['Date'].unique()

airlines['Date'] = airlines['Date'].apply(lambda x: x.split(' ')[0])
airlines['Date'] = pd.to_datetime(airlines['Date'], format='%d/%m/%Y')#dayfirst=True)

#El tiempo total desde el rodaje por pista salida/llegada + en el aire.
airlines['TiempoTotal']=airlines['AirTime']+airlines['TaxiIn']+airlines['TaxiOut']
airlines['TaxiTotal']=airlines['TaxiIn']+airlines['TaxiOut']

# Eliminamos los campos de fecha

airlines.drop(['Unnamed: 0','DayofMonth', 'Month','Year'], axis = 'columns', inplace=True)


```

***2.3 Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).***

```python
# Fórmula de velocidad media = distancia / Tiempo
# No tenemos en cuenta el tiempo en rodajeen tierra TaxiIn/Out
# Creamos la columna VelocidadMedia
# unidad millas/minutos

airlines['VelocidadMedia']=airlines['Distance']/(airlines['AirTime'])

airlines.head()

#Otro punto para estudiar
#print(f'- Velocidad máxima: {airlines["VelocidadMedia"].max()}')

# Creamos la columna Retraso
# Utilizamos el estado booleano si es Delay = True, si no False
airlines['Retraso'] =(airlines['ArrDelay'] > 0) | (airlines['DepDelay'] > 0)
airlines.head()
```
